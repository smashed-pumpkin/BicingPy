### **Objective of the script**: 
**Identify the best time to pick up and drop off a bike from stations nearby**

#### 1. Explore the data

Load the dataset (should be stored locally):

```python
import sqlite3
conn = sqlite3.connect("bicidb.db")
cursos = conn.cursor()

import pandas as pd
df = pd.read_sql("SELECT * FROM historical_availability;", con=conn)
```

Take a first glance at the structure. We should have one observation for all bike locations (lat and lon) per date and time. Each observation contains the following information: 

 * **Station Name** - the name of the station, which corresponds to the street name and number on which it is located
 * **Station Lat** - the latitude of the station
 * **Station Lon** - the longitude of the station
 * **Station ID** - the unique ID of the station
 * **Bikes** - the number of available bikes at that time
 * **Bikes OK** - a flag that indicated whether there are more than 1 bikes at that time (1=Y, 0=N)
 * **Slots** - the number of available bike slots at that time
 * **Slots OK** - a flag that indicated whether there are more than 1 slots at that time (1=Y, 0=N)
 * **Entry Date** - the time at which the data was downloaded

```python
df.head(5)
```

Break out the entry date into separate date/time components to make it easier for analysis and exploration:

```python
df['entry_date'] = pd.to_datetime(df['entry_date'])
df['entry_year'] = [x.year for x in df['entry_date']]
df['entry_month'] = [x.month for x in df['entry_date']]
df['entry_day'] = [x.day for x in df['entry_date']]
df['entry_hour'] = [x.hour for x in df['entry_date']]
df['entry_minute'] = [x.minute for x in df['entry_date']]
df['entry_weekday'] = [x.weekday() for x in df['entry_date']]
df['entry_weekdayname'] = [x.strftime("%a") for x in df['entry_date']]
df.head(3)
```

Since the script has been run failry sporadically and we are interested in identifying the right time to pick up/drop off a bike, let's check how many observation are available per hour:

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

num_bins = len(df.entry_hour.unique())
n, bins, patches = plt.hist(df.entry_hour, num_bins, facecolor='green', alpha=0.5) # the alpha makes the colour lighter
plt.show()
```

We then use a range of X meters to narrow down the dataset to only local stations. 

Note that we are using another bicing API which tells us the nearest within the specified distance stations.

```python
import json
import requests

current_lat = current_latlon[0]
current_lon = current_latlon[1]
distance = 0.3 # in kilometers

url = 'http://barcelonaapi.marcpous.com/bicing'
url_nearest = url+'/nearstation/latlon/'+str(current_lat)+'/'+str(current_lon)+'/'+str(distance)+'.json'
res_nearest = requests.get(url_nearest)
json_nearby = json.loads(res_nearest.text)
```

```python
station_id_nearby = []

for idx,i in enumerate(json_nearby['data']['nearstations']):
    station_id_nearby.append(int(json_nearby['data']['nearstations'][idx]['id']))
```

```python
df_filtered = df[df.station_id.isin(station_id_nearby)].groupby(['entry_weekdayname','entry_hour'])['bikes'].mean()
df_filtered = df_filtered.reset_index(name="means")

# Sort the data in the correct weekday order
df_filtered['entry_weekdayname'] = pd.Categorical(df_filtered['entry_weekdayname'], ['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])
df_filtered = df_filtered.sort_values(by='entry_weekdayname')

# Pivot the dataframe to create a [hour x weekday] matrix containing means
df_filtered_pivot = df_filtered.pivot('entry_hour','entry_weekdayname','means')

import seaborn as sns #visualization library based on matplotlib
sns.heatmap(df_filtered_pivot, annot=False, cmap="PuBuGn")
plt.show()
```

#### 3. Map out the data (static view)

Create a subset of the dataset that calculates the average available slots / bikes for each station_lat,  station_lon, station_name:

```python
df_avg_bikes = pd.DataFrame(df.groupby(['station_lat', 'station_lon', 'station_name'])['bikes'].mean().reset_index(name='bikes'))
df_avg_bikes
```

```python
df_avg_slots = pd.DataFrame(df.groupby(['station_lat', 'station_lon', 'station_name'])['slots'].mean().reset_index(name='slots'))
df_avg_slots
```

Join the two avg data sets:

```python
df_avg = pd.concat([df_avg_bikes, df_avg_slots.slots], axis=1)
df_avg
```

Create a function that locates the respective files (those need to be saved in the Credentials folder):

```python
from pathlib import Path
import os

def locate(filename):
    cwd = Path(os.getcwd())
    cwd_new = str(cwd.parent)+'\Credentials'
        
    for r,d,f in os.walk(cwd_new):
        for files in f:
            if files == filename:
                path = os.path.join(r,files)
              
    with open(path, 'r') as f:
        content = f.readlines()
        content = [x.strip() for x in content]
        
    return(content)
```

Get your current location (needed for folium to know what parts of the map to show):

```python
from geopy.geocoders import GoogleV3

address = locate('home_address.txt')[0]
key = locate('login_googleAPI.txt')[0]

geolocator = GoogleV3(api_key=key)
current_latlon = geolocator.geocode(address)[1]
```

Select the lat/lon you want to default the map on and how zoomed you want it to be:

```python
import folium

folmap_bikes = folium.Map(location=current_latlon, tiles='OpenStreetMap', zoom_start=20)
folmap_slots = folium.Map(location=current_latlon, tiles='OpenStreetMap', zoom_start=20)
```

Add to the map the markers of the available stations:

```python
for idx,i in enumerate(df_avg.station_name):
    popup = folium.Popup(df_avg.station_name[idx]+' / Available bikes: '+str(int(df_avg.bikes[idx]))+' / Available slots: '+str(int(df_avg.slots[idx])), parse_html=True)
    folium.Marker([df_avg.station_lat[idx], df_avg.station_lon[idx]], popup = popup
                 ).add_to(folmap_bikes)
```
